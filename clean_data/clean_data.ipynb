```python
import pandas as pd
```

```python
vol_raw = "../raw_data/vol_data.csv"
invol_raw = "../raw_data/invol_data.csv"

vol_df = pd.read_csv(vol_raw)
invol_df = pd.read_csv(invol_raw)
```

```python
vol_df
```

```python
invol_df
```

```python
vol_df["Order Type"] = "Voluntary"

del vol_df["Unnamed: 5"]
del vol_df["Docket#"]

vol_df["Received"] = pd.to_datetime(vol_df["Received"])
vol_df["Resolved"] = pd.to_datetime(vol_df["Resolved"])

vol_df
```

```python
invol_df["Order Type"] = "Involuntary"

del invol_df["City"]

invol_df["Year Placed"] = invol_df["Year Placed"].fillna(0).astype(int)
invol_df["Year Placed"] = invol_df["Year Placed"].astype(str)
invol_df["Year Placed"] = invol_df["Year Placed"].replace(['0'],'Unknown year')

invol_df
```

```python
vol_df["Region"] = ""

invol_df["Region"] = ""
```

```python
vol_df
```

```python
invol_df
```

```python
vol_df.to_csv("clean_vol_data.csv")

invol_df.to_csv("clean_invol_data.csv")
```
