## State of Slavery Before & During the Civil War

### Slave State (Confederacy - 11): TX, LA, AR, TN, MS, AL, GA, FL, SC, NC, VA
### Border State (Had Slavery but part of the Union - 5): MO, KY, WV, MD, DE
### Free State (Union - 20): CA, OR, KS, IA, MN, WI, IL, IN, MI, OH, PA, NY, NJ, CT, RI, MA, NH, VT, ME, NV
